## Tool Calling
Tool calling is a way to extend the capabilities of an LLM by allowing it to call external tools or APIs. This can be useful for tasks that require access to external data or services.

```python
!pip install requests
```

```python
from agents import Agent, function_tool
import requests

@function_tool
def get_weather(city: str) -> str:
    base_url = "http://api.openweathermap.org/data/2.5/weather?"
    complete_url = f"{base_url}q={city}&appid={openweather_key}"

    response = requests.get(complete_url)
    data = response.json()

    weather_data = data["weather"][0]
    weather_description = weather_data["description"]
    return f"The current weather in {city} is {weather_description}."

@function_tool
def get_temperature(city: str) -> str:
    base_url = "http://api.openweathermap.org/data/2.5/weather?"
    complete_url = f"{base_url}q={city}&appid={openweather_key}&units=metric"

    response = requests.get(complete_url)
    data = response.json()

    main_data = data["main"]
    temperature = main_data["temp"]
    return f"The current temperature in {city} is {temperature}°C."

agent = Agent(
    name="Weather Agent",
    model="gpt-4.1-mini",
    instructions="""You are the local weather agent. You are given a city and need to tell both the weather and temperature.
    For any unrelated queries, say I can't help with that. Always use the provided tools to get weather and temperature information.""",
    tools=[get_weather, get_temperature]
)

result = await Runner.run(agent, "What is the weather in Singapore")
print(result.final_output)
```

```python
from agents import Agent, function_tool
from tavily import TavilyClient

@function_tool
def tavily_search(query: str) -> str:
    tavily = TavilyClient(api_key=tavily_key)
    response = tavily.search(query=query, search_depth="basic")

    results = response.get('results', [])
    summary = "\n".join([f"Source: {res['url']}\nContent: {res['content']}" for res in results])
    return summary

search_agent = Agent(
    name="Search Agent",
    model="gpt-4.1-mini",
    instructions="You are a search agent. You will provide the answer after searching the internet using the Tavily API.",
    tools=[tavily_search]
)

result = await Runner.run(search_agent, "Who is the Prime Minister of Singapore")
print(result.final_output)
```

```python
from agents import Agent, Runner, handoff, RunContextWrapper

history_tutor_agent = Agent(
    name="History Tutor",
    model="gpt-4.1-mini",
    handoff_description="Specialist agent for historical questions",
    instructions="You provide assistance with historical queries. Explain important events and context clearly.",
)

math_tutor_agent = Agent(
    name="Math Tutor",
    model="gpt-4.1-mini",
    handoff_description="Specialist agent for math questions",
    instructions="You provide assistance with math queries. Explain your reasoning at each step and include examples"
)

physics_tutor_agent = Agent(
    name="Physics Tutor",
    model="gpt-4.1-mini",
    handoff_description="Specialist agent for physics questions",
    instructions="You provide assistance with physics queries. Explain concepts, formulas, and problem-solving steps clearly."
)

def on_math_handoff(ctx: RunContextWrapper[None]):
    print("Handing off to math tutor agent")

def on_history_handoff(ctx: RunContextWrapper[None]):
    print("Handing off to history tutor agent")

def on_physics_handoff(ctx: RunContextWrapper[None]):
    print("Handing off to physics tutor agent")

# This agent has the capability to handoff to either the history or math tutor agent
triage_agent = Agent(
    name="Triage Agent",
    instructions="""
You are a homework assistant responsible for routing user questions to the most appropriate tutor agent.

## Step 1: Identify the subject of the user’s homework question.
Determine the primary academic domain the question belongs to.

## Step 2: Select the correct agent based on the criteria below:

### Physics Tutor Agent
Route the question to the Physics Tutor Agent if it involves:
Physical laws or principles (e.g., motion, forces, energy, electricity, magnetism, thermodynamics, optics)
Physics formulas, units, experiments, or conceptual explanations
Problem-solving that relies on physics concepts, even if math is involved

### Math Tutor Agent
Route the question to the Math Tutor Agent if it involves:
Pure mathematical concepts or procedures (e.g., algebra, geometry, trigonometry, calculus, statistics)
Solving equations, proofs, word problems, or numerical reasoning
Math techniques without requiring domain-specific physics context

### History Tutor Agent
Route the question to the History Tutor Agent if it involves:
Historical events, dates, timelines, or periods
Historical figures, movements, wars, revolutions, or civilizations
Analysis of historical causes, consequences, or interpretations

## Step 3: Handle mixed or unclear cases.

If one subject clearly dominates, route to that agent.
If the question genuinely spans multiple subjects, choose the agent most essential to solving the problem.
If no tutor agent is clearly relevant, do not hand off—provide a general response yourself.

## Step 4: Default behavior.
If the question is not homework-related or does not match physics, math, or history, respond directly with a general explanation or assistance.

""",
    handoffs=[handoff(history_tutor_agent, on_handoff=on_history_handoff),
              handoff(math_tutor_agent, on_handoff=on_math_handoff),
              handoff(physics_tutor_agent, on_handoff=on_physics_handoff)]
)
```

## Agent Session/Memory

When session memory is enabled:

1. Before each run: The runner automatically retrieves the conversation history
for the session and prepends it to the input items.
2. After each run: All new items generated during the run (user input, assistant responses, tool calls, etc.) are automatically stored in the session.
3. Context preservation: Each subsequent run with the same session includes the full conversation history, allowing the agent to maintain context.

```python
from agents import Agent, Runner, SQLiteSession

# Create agent
agent = Agent(
    name="Assistant",
    model="gpt-4.1-mini",
    instructions="Reply very concisely.",
)

# Create a session instance with a session ID
session = SQLiteSession("conversation_123")
```

```python
# First turn
result = await Runner.run(
    agent,
    "What city is the Golden Gate Bridge in?",
    session=session
)
print(result.final_output)
```

```python
# Second turn - agent automatically remembers previous context
result = await Runner.run(
    agent,
    "What state is it in?",
    session=session
)
print(result.final_output)
```

```python
result = await Runner.run(
    agent,
    "What's the population of the state?",
    session=session
)
print(result.final_output)
```

## Guardrails
Guardrails are a way to validate the input and output of an Agent to insure proper usage.

```python
from agents import Agent, Runner, ModelSettings

agent=Agent(
    name="MyAgent",
    model="gpt-4o",
    instructions="You are helpful assistant",
    model_settings=ModelSettings(
        temperature=0.5,
        max_tokens=150,
        tool_choice="auto",
        top_p=1,
        top_k=40,
        verbose=False
    ),
    input_guardrails=[],
    output_guardrails=[],
    tools=[],
    handoffs=[],
)
```

```python
from agents import Agent, GuardrailFunctionOutput, RunContextWrapper, Runner, TResponseInputItem, input_guardrail
from pydantic import BaseModel

class HomeworkCheatDetectionOutput(BaseModel):
    attempting_cheat: bool
    explanation: str

homework_cheat_guardrail_agent = Agent(
    name="Homework Cheat Detector",
    instructions=(
        "Determine if the user's query resembles a typical homework assignment or exam question, indicating an attempt to cheat. General questions about concepts are acceptable. "
        " Cheating: 'Fill in the blank: The capital of France is ____.',"
        " 'Which of the following best describes photosynthesis? A) Cellular respiration B) Conversion of light energy C) Evaporation D) Fermentation.'"
        " Not-Cheating: 'What is the capital of France?', 'Explain photosynthesis.'"
    ),
    output_type=HomeworkCheatDetectionOutput,
    model="gpt-4o-mini"
)

@input_guardrail
async def cheat_detection_guardrail(
        ctx: RunContextWrapper[None], agent: Agent, input: str | list[TResponseInputItem]
) -> GuardrailFunctionOutput :

    detection_result = await Runner.run(homework_cheat_guardrail_agent, input)

    return GuardrailFunctionOutput(
        tripwire_triggered=detection_result.final_output.attempting_cheat,
        output_info=detection_result.final_output
    )

study_helper_agent = Agent(
    name="Study Helper Agent",
    instructions="You assist users in studying by explaining concepts or providing guidance, without directly solving homework or test questions.",
    model="gpt-4o",
    input_guardrails=[cheat_detection_guardrail]
)
```

```python
# This should trigger the cheat detection guardrail
from agents import InputGuardrailTripwireTriggered

try:
    response = await Runner.run(study_helper_agent, "Fill in the blank: The process of converting light energy into chemical energy is called ____.")
    print("Guardrail didn't trigger")
    print("Response: ", response.final_output)

except InputGuardrailTripwireTriggered as e:
    print("Homework cheat guardrail triggered")
    print("Exception details:", str(e))
```

## Upload PDF to RAG

```python
from pypdf import PdfReader
from pathlib import Path
from openai import OpenAI
from agents import Agent, function_tool

import chromadb
chroma = chromadb.Client()

# Clear existing collection and then get it to ensure a fresh start
if "pdf_rag_kb" in [c.name for c in chroma.list_collections()]:
    chroma.delete_collection(name="pdf_rag_kb")
collection = chroma.get_or_create_collection(name="pdf_rag_kb")

def load_pdf_text(pdf_path: str) -> str:
    reader = PdfReader(pdf_path)
    pages = []
    for i, page in enumerate(reader.pages):
        text = page.extract_text()
        if text:
            pages.append(f"[Page {i+1}]\n{text}")
    return "\n".join(pages)

def chunk_text(text, chunk_size=800, overlap=120):
    chunks = []
    start = 0
    while start < len(text):
        end = start + chunk_size
        chunks.append(text[start:end])
        start = end - overlap
    return chunks

def index_pdfs(pdf_dir: str):
    pdf_dir_path = Path(pdf_dir)

    if not pdf_dir_path.exists():
        print(f"Warning: PDF directory '{pdf_dir}' does not exist. Please create it and add PDFs.")
        return

    pdf_files = list(pdf_dir_path.glob("*.pdf"))

    if not pdf_files:
        print(f"Warning: No PDF files found in '{pdf_dir}'. No PDFs indexed.")
        return

    all_chunks = []
    metadatas = []
    ids = []

    for pdf in pdf_files:
        text = load_pdf_text(str(pdf))
        chunks = chunk_text(text)

        for i, chunk in enumerate(chunks):
            if chunk.strip(): # Only add non-empty chunks
                all_chunks.append(chunk)
                metadatas.append({
                    "source": pdf.name,
                    "chunk": i
                })
                ids.append(f"{pdf.stem}_{i}")

    if not all_chunks:
        print("Warning: No valid text chunks extracted from PDFs. Skipping embedding and upload.")
        return

    embeddings = client.embeddings.create(
        model="text-embedding-3-small",
        input=all_chunks
    ).data

    collection.add(
        documents=all_chunks,
        metadatas=metadatas,
        embeddings=[e.embedding for e in embeddings],
        ids=ids
    )
    print(f"Successfully indexed {len(all_chunks)} chunks from {len(pdf_files)} PDF(s).")

client = OpenAI()

# Always re-index after clearing
index_pdfs("./fake_invoices")

@function_tool
def pdf_retrieve_context(question: str) -> str:
    q_embedding = client.embeddings.create(
        model="text-embedding-3-small",
        input=question
    ).data[0].embedding

    results = collection.query(
        query_embeddings=[q_embedding],
        n_results=4
    )

    return "\n\n".join(results["documents"][0])

agent = Agent(
    name="Assistant",
    model="gpt-4.1-mini",
    instructions="""
You are a helpful assistant.
Always call pdf_retrieve_context before answering.
Use the retrieved PDF content to answer accurately.
If the PDFs do not contain the answer, say so clearly.
""",
    tools=[pdf_retrieve_context],
)
```

```python
result = await Runner.run(
    agent,
    "Which invoices sell Samsung product?"
)
print(result.final_output)
```
